```python
pbp = nfl.import_pbp_data([2024])
```

```python
#further identifying pbp data
# Check what the first few rows look like for a single play
pbp.iloc[0]

```

```python
#397 columns is a lot, to make easier we partition based on data availability
null_pct = pbp.isnull().mean().sort_values(ascending=False)
print(f"Columns with >90% null: {(null_pct > 0.90).sum()}")
print(f"Columns with >50% null: {(null_pct > 0.50).sum()}")
print(f"Columns with <10% null: {(null_pct < 0.10).sum()}")
```

```python
# Columns that are >90% null — candidates to drop
null_pct[null_pct > 0.90].index.tolist()
```

```python
# Get columns with <10% null, in their original order from the dataset
low_null_cols = [col for col in pbp.columns if null_pct[col] < 0.10]

with open('/Users/anthonyjaramillo/PycharmProjects/4900/data/pbp_columns.txt', 'w') as f:
    for i, col in enumerate(low_null_cols, 1):
        f.write(f"{i}. {col}\n")

print(f"Saved {len(low_null_cols)} columns to data/pbp_columns.txt")
```

```python
# Get first row's values, dtypes, and column names
first_row = pbp.iloc[0]
dtypes = pbp.dtypes

# Build a summary table
summary = pd.DataFrame({
    'column_name': pbp.columns,
    'dtype': [str(dtypes[col]) for col in pbp.columns],
    'example_value': [first_row[col] for col in pbp.columns]
})

# Only keep columns with <10% null
low_null_cols = [col for col in pbp.columns if null_pct[col] < 0.10]
summary = summary[summary['column_name'].isin(low_null_cols)].reset_index(drop=True)
summary.index = summary.index + 1  # Start numbering at 1

# Save to file
summary.to_csv('/Users/anthonyjaramillo/PycharmProjects/4900/data/pbp_column_details.csv', index=True)

print(f"Saved {len(summary)} columns to data/pbp_column_details.csv")
summary
```
